### Visualize the data

```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
#headers = ['country', 'gdpPercap_1952', 'gdpPercap_1957', 'gdpPercap_1962', 'gdpPercap_1967', 'gdpPercap_1972', 'gdpPercap_1977', 'gdpPercap_1982', 'gdpPercap_1987', 'gdpPercap_1992', 'gdpPercap_1997', 'gdpPercap_2002', 'gdpPercap_2007']
#df = pd.read_csv('data/gapminder_gdp_europe.csv', usecols=headers)

# read the data into a dataframe (dfd)

dfd = pd.read_csv('data/gapminder_gdp_europe.csv', index_col = 'country')

# transpose it to have the countries as rows

dfd = dfd.T

# rename to make it look nice in the plot

df = dfd.rename(index = {'gdpPercap_1952':'1952','gdpPercap_1957':'1957','gdpPercap_1962':'1962','gdpPercap_1967':'1967','gdpPercap_1972':'1972','gdpPercap_1977':'1977','gdpPercap_1982':'1982','gdpPercap_1987':'1987','gdpPercap_1992':'1992','gdpPercap_1997':'1997','gdpPercap_2002':'2002','gdpPercap_2007':'2007'})

# print your data frame to check

#print("Contents in csv file:", df)


# plot selected countries (select more of less bu adding more lines; plotting all would be too much)

df.Switzerland.plot(legend=True) #plot Switzerland column
df.Germany.plot(legend=True)
df.France.plot(legend=True)
df.Greece.plot(legend=True)
df.Italy.plot(legend=True)
df.Norway.plot(legend=True)
df.Montenegro.plot(legend=True)
plt.title('gdp per capita')
plt.show()
```
